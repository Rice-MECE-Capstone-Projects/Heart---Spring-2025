
# Wavelet-based PCG denoising

```python
import wave
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt
import pywt
```

```python
def load_wav(wav_path):
    samplerate, data = wavfile.read(wav_path)
    #start=int(len(data)*0.1)
    #end=int(len(data)*0.9)
    #data=data[start:end]
    data=data/np.max(data)
    return samplerate, data
```

```python
# denoising noisy signal
noisy_path='./data/noisy_pcg/23616/23616_MV.wav'
# denoising clean signal+noise model
clean_path = './data/clean_pcg/New_N_001.wav'
```

```python

samplerate, data = load_wav(clean_path)

plt.figure(figsize=(10, 4))
plt.plot(np.linspace(0, len(data)/samplerate, len(data)), data, color='blue')
plt.title('Original PCG signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
print("auscultation point:")
print('PCG signal sample rate:', samplerate)
print('PCG signal length:',f"{len(data)/samplerate:.2f}s")
print('murmur:')
print('murmur type:')
```

```python
def get_signal_power(s):
    return np.mean(s**2)
def get_snr(signal, noise):
    powS = get_signal_power(signal)
    powN = get_signal_power(noise)
    return 10*np.log10((powS-powN)/powN)
```

```python
# constructing noise model
# constructing noise model based on certain SNR level
target_snr_db = 10
# red noise
red_spec=lambda f:1 / np.where(f == 0, float('inf'), f)
# pink noise
pink_spec=lambda f: 1 / np.where(f == 0, float('inf'), np.sqrt(f))
N=data.shape[0]
def get_noise_model(func,N,target_snr):
    signal_power = get_signal_power(data)
    sig_avg_db = 10 * np.log10(signal_power)
    noise_avg_db = sig_avg_db - target_snr_db
    noise_avg_watts = 10 ** (noise_avg_db / 10)
    noise_volts = np.random.normal(0, np.sqrt(noise_avg_watts), N)
    white_noise=np.fft.rfft(noise_volts)
    S=func(np.fft.rfftfreq(N))
    S=S/np.sqrt(np.mean(S**2) + 1e-10)
    return np.fft.irfft(white_noise*S)


redNoise=get_noise_model(red_spec,N,target_snr_db)
pinkNoise=get_noise_model(pink_spec,N,target_snr_db)


f=np.fft.rfftfreq(N,d=1/samplerate)
get_psd=lambda f: 10*np.log10(np.abs(np.fft.rfft(f))/np.sqrt(samplerate*N))
plt.figure(figsize=(14, 8))
plt.subplot(2, 1, 1)
plt.plot(f,get_psd(pinkNoise), color='pink')
plt.title("Pink Noise (1/f Noise)")
plt.grid()
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency(dB/Hz)',fontsize=16)
plt.ylim(bottom=-50)

# Subplot for Red Noise
plt.subplot(2, 1, 2)
plt.plot(f,get_psd(pinkNoise), color='red')
plt.title("Red Noise (Brownian Noise)")
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power/Frequency(dB/Hz)',fontsize=16)
plt.grid()

plt.tight_layout()
plt.ylim(bottom=-50)
plt.show()
```

```python
noised_data=data+np.pad(pinkNoise,(0,len(data)-len(pinkNoise)),'constant',constant_values=0)
plt.figure(figsize=(12, 4))
plt.plot(np.linspace(0, len(noised_data)/samplerate, len(data)), noised_data, color='blue')
plt.title('Original PCG signal added with noise')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```

```python
mother_wavelet=pywt.Wavelet('coif5')
level=6
coeffs = pywt.wavedec(noised_data, mother_wavelet, level=level,mode='symmetric')
```

```python
fig, axs = plt.subplots(len(coeffs), 1, figsize=(10, 12))

for i, coeff in enumerate(coeffs):
    axs[i].plot(coeff)
    if i == 0:
        axs[i].set_title(f"Approximation Coefficients", fontsize=12)
    else:
        axs[i].set_title(f"Detail Coefficients (Level {level+1-i})", fontsize=12)
    axs[i].set_xlim(0, len(coeff))
fig.suptitle(f"Wavelet Decomposition using coif5", fontsize=16)
plt.tight_layout()
plt.show()
print("Frequency band of different levels:")
fre_high=samplerate//2
fre_low=samplerate//4
for i in range(1,level+1):
    print(f"Level {i}: {fre_low}-{fre_high}Hz")
    fre_low=fre_low//2
    fre_high=fre_high//2
print(f"Approximation level: 0-{fre_high}Hz")
```

```python
def get_med_75(arr):
    index_75=int(0.65*(len(arr)-1))
    arr_sorted=sorted(np.abs(arr))
    return arr_sorted[index_75]
```

```python
# Threshold
flag=3
coeff_m=np.array([np.mean(np.abs(coeffs[i])) for i in range(flag)])
coeff_v=np.array([np.var(np.abs(coeffs[i])) for i in range(flag)])
coeff_med75=np.array([get_med_75(coeffs[i]) for i in range(flag)])
coeff_th=np.empty(flag)
for i in range(flag):
    if coeff_v[i]>coeff_med75[i]:
        coeff_th[i]=coeff_med75[i]*(1-(coeff_v[i]-coeff_med75[i]))
    elif coeff_v[i] < coeff_med75[i] < coeff_m[i]:
        coeff_th[i]=coeff_med75[i]
    else:
        coeff_th[i]=coeff_med75[i]+coeff_med75[i]-coeff_m[i]
print(coeff_th)
```

```python
# Threshold rescale using mln
for i in range(flag):
    coeff_th[i]=coeff_th[i]*np.median(np.abs(coeffs[i]))/0.6745
```

```python
def non_linear_th(d,th,alpha,beta):
    T1=th*alpha
    T2=th*beta
    if np.abs(d)<T1:
        return 0
    elif T1 < np.abs(d) < T2:
        return pow(d,3)/pow(T2,2)
    else:
        return d
```

```python
# applying threshold to coefficients
coeffs_denoised=[np.empty(len(coeffs[i])) for i in range(len(coeffs))]
beta_1=1.3
beta_2=1.4
for i in range(len(coeffs)):
    if i>=flag:
        coeffs_denoised[i]=np.zeros(len(coeffs[i]))
    elif coeff_v[i]>coeff_med75[i]:
        arr=np.array([non_linear_th(d,coeff_th[i],1,beta_2) for d in coeffs[i]])
        coeffs_denoised[i]=arr
    else:
        arr=np.array([non_linear_th(d,coeff_th[i],1,beta_1) for d in coeffs[i]])
        coeffs_denoised[i]=arr
```

```python
# Reconstruct the signal
denoised_data=pywt.waverec(coeffs=coeffs_denoised, wavelet=mother_wavelet)
plt.figure(figsize=(10, 10))
plt.subplot(2, 1, 2)
plt.plot(np.linspace(0, len(denoised_data)/samplerate, len(denoised_data)), denoised_data, color='blue')
plt.title('Denoised PCG signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.subplot(2, 1, 1)
plt.plot(np.linspace(0, len(data)/samplerate, len(data)), noised_data, color='blue')
plt.title('Original PCG signal')
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.show()
```
